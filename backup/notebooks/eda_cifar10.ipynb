```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X_train = np.load("../data/processed/cifar10/X_train.npy")
y_train = np.load("../data/processed/cifar10/y_train.npy")

X_test = np.load("../data/processed/cifar10/X_test.npy")
y_test = np.load("../data/processed/cifar10/y_test.npy")

print("Train shape:", X_train.shape, y_train.shape)
print("Test shape:", X_test.shape, y_test.shape)
```

```python
print("NaNs in X_train:", np.isnan(X_train).sum())
print("NaNs in y_train:", np.isnan(y_train).sum())
```

```python
class_counts = np.bincount(y_train)

plt.figure(figsize=(8, 4))
sns.barplot(x=list(range(10)), y=class_counts)
plt.xlabel("Class Label")
plt.ylabel("Number of Samples")
plt.title("Class Distribution in CIFAR-10 (Training Set)")
plt.show()
```

```python
class_names = [
    "airplane", "automobile", "bird", "cat", "deer",
    "dog", "frog", "horse", "ship", "truck"
]

plt.figure(figsize=(10, 6))
for i in range(10):
    idx = np.where(y_train == i)[0][0]
    plt.subplot(2, 5, i + 1)
    plt.imshow(X_train[idx])
    plt.title(class_names[i])
    plt.axis("off")

plt.suptitle("Sample Images from Each CIFAR-10 Class")
plt.show()
```

```python
pixels = X_train.flatten()

plt.figure(figsize=(6, 4))
plt.hist(pixels, bins=50)
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.title("Pixel Intensity Distribution")
plt.show()
```
